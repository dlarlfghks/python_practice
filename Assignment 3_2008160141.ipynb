◾2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)◦>>> addallodd([1])

◦>>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])

◾[참고]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.) 


```python
def addall(inputnum):
    c = 0
    for x in inputnum:
        c += int(x)
    return c

def addallodd(inputnum):
    c = 0
    for x in inputnum:
        if int(x)%2:
            c += int(x)
    return c

a = raw_input(">>>")
a = a[:a.index('(')],a[a.index('[')+1:a.index(')')-1]
a = list(a)
a[1] = a[1].split(',')

if a[0] == "addall":
    print addall(a[1])

elif a[0] == "addallodd":
    print addallodd(a[1])
```

이 때 10,001번째의 소수를 구하세요.

```python
number = input('>>>')
count = 0
num = 1

def primenum(n):
    if n > 1:
        for i in range(2,n):
            if (n % i) == 0:
                return False
        return True

while 1:
    num += 1
    if primenum(num):
        count += 1
        if count == number:
            print num
            break
```

◾요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.◦예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']

◾요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.◦예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}

◾요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
◦http://www.naver.com
◦http://www.daum.net
◦http://www.nytimes.com


```python
#-*- coding: utf-8 -*-
import urllib2
import string

def char(url):
    source = urllib2.urlopen(url).read()
    words = []

    for i in range(0, source.count('<!--')):
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    for i in range(0, source.count("<")):
        if source.find("<") != -1:
            source = source.replace(source[source.find("<"):source.find(">")+1:],"")

    source = source.split()  

    for word in source:
        word = word.translate(None, string.punctuation)
        if word == '':
            pass
        else :
            words.append(word.translate(None, string.punctuation))

    clone = words
    dic = dict()

    for word in words:
        count = 0
        for element in clone:
            if word == element :
                count+=1
        dic.update({word : count})

    print dic.items()
    
char("http://www.naver.com")
print
char("http://www.daum.net")
print
char("http://www.nytimes.com")
```
